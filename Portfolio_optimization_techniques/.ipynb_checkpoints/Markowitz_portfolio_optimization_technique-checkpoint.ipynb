# RETURNS 

$$ r_t = \frac{p_t}{p_{t-1}} $$

```python
returns = stocks / stocks.shift(1)
returns 
```

# LOG RETURNS 
## It is a function that favours or promotes fairness among all the stocks


$$ \log(r_t) = \log\left(\frac{p_t}{p_{t-1}}\right) $$

```python
import numpy as np
logReturns = np.log(returns)
logReturns 
```

1. **Excess Return**:
   $$ R(w) - R_g $$
   - Where \( R(w) \) is the return of the portfolio with weights \( w \), and \( R_g \) is the return of the benchmark or risk-free rate.

2. **Sharpe Ratio**:
   $$ SR(U) = \frac{R(U) - R_f}{\sigma(U)} $$
   - Where \( SR(U) \) is the Sharpe Ratio of the portfolio \( U \), \( R(U) \) is the return of the portfolio, \( R_f \) is the risk-free rate, and \( \sigma(U) \) is the standard deviation of the portfolio's returns.

3. **Weights**:
   $$ W = [w_1, w_2, w_3, \ldots, W_k] $$
   - This indicates a set of weights \( W \) assigned to different assets in the portfolio, where \( W_k = 1 \) signifies that the sum of all weights equals 1.

4. **Weight Constraint**:
   $$ \sum_{k=1}^{n} W_k = 1 $$
   - This formula states that the sum of the weights \( W_k \) for all assets \( k \) in the portfolio must equal 1, ensuring that the entire portfolio is fully invested.

- Since we willlook for max $$SR(w)$$ and $$R_f$$ is common for all $$w's$$,then we shall compute 

$$ R(w) = {w} ^ {T} log(r) $$
- This is how we calculate the log asset returns 

### Volatility Per Portfolio: It is the square root of the variance 
$$ \sigma(w) = \sqrt{w^T \Sigma w} $$

### Sharpe Ratio:
$$ SR(w) \sim \frac{w^T \log(r)}{\sqrt{w^T \Sigma w}} $$

We're looking for a 'w' which maximizes the sharpe ratio

# Return vs Volatility Chart
Generating (almost) all possible portfolios

```python
noOfPortfolios = 10000
weight = np.zeros((noOfPortfolios,4))
expectedReturn = np.zeros(noOfPortfolios)
expectedVolatility = np.zeros(noOfPortfolios)
sharpeRatio = np.zeros(noOfPortfolios)
meanLogRet = logReturns.mean()
Sigma = logReturns.cov()
for k in range(noOfPortfolios):
    #generate random weights vector 
    w = np.array(np.random.random(4))
    w = w / np.sum(w)
    weight[k,:] = w
    #Expected Log return
    expectedReturn[k] = np.sum(meanLogRet * w)
    #Expected volatility
    expectedVolatility[k] =np.sqrt( np.dot(w.T, np.dot(Sigma, w)) )
    #Sharpe Ratio
    sharpeRatio[k] = expectedReturn[k]/expectedVolatility[k]
```

```python
import matplotlib.pyplot as plt
plt.plot(expectedReturn)
```

# Sharpe Ratio Maximization

```python
maxIndex = sharpeRatio.argmax() #gives us the index of the max sharpe ratio 
weight[maxIndex,:] #getting the weights at this location
#so if you've got thousand dollars in your portfolio then you should allocate:
# $1  - CEVA, $3 Cents - Google, $345 - Tesla, $649 - Zoomedica
```

## SCATTER PLOT: RETURNS vs VOLATILITY
#### RETURNS - y-axis, VOLATILITY - x-axis

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,6))
plt.scatter(expectedVolatility, expectedReturn, c=sharpeRatio)
plt.xlabel('Expected Volatility')
plt.ylabel('Expected log Returns')
plt.colorbar(label='Sharpe Ratio')
plt.scatter(expectedVolatility[maxIndex],expectedReturn[maxIndex],c='red')

plt.show()
```

# Efficient Markowitz Frontier
## Maximizing the return and at the same time minimizing the risk
Minimize \($${w}$$\):
$$ w^T \Sigma x $$

Subject to:
$$ \bar{p}^{-T} w \geq r_{min} $$
$$ w \geq 0 $$
$$ 1^T w = 1 $$

## Maximize the Sharpe Ratio or minimize it's negative with subject to the following constraints: 
Minimize \( w \):
$$ -SR(w) $$

Subject to:
$$ 0 \leq w_i \leq 1, \quad i = 1, \ldots, 4 $$

```python
from scipy.optimize import minimize

def negativeSR(w):
    w = np.array(w)
    R = np.sum(meanLogRet*w) #expected returns
    V = np.sqrt( np.dot(w.T, np.dot(Sigma, w)) )
    SR = R/V
    return -1*SR

def checkSumToOne(w):
    return np.sum(w)-1 #if 1 then it returns 0
    
w0 = [0.25,0.25,0.25,0.25] #equiprobable
bounds = ((0,1),(0,1),(0,1),(0,1))
constraints = ({'type':'eq', 'fun':checkSumToOne}) #minimize function returns w vector and now we are going to check each solution if it sums upto 1 or not 
w_opt = minimize(negativeSR,w0,method='SLSQP',bounds=bounds,constraints=constraints)
w_opt

#the x which gets printed out is the accurate solution
```

```python
#solution
w_opt.x = np.round(w_opt.x, decimals=10)
w_opt.x
```

```python
#for each horizontal line on the plot - return, we need to find the best volatility and save all such points and then plot this frontier
returns = np.linspace(0,0.07,50)
volatility_opt = []
def minimizeMyVolatility(w):
    w = np.array(w)
    V = np.sqrt( np.dot(w.T, np.dot(Sigma, w)) )
    return V
def getReturn(w):
    w = np.array(w)
    R = np.sum(meanLogRet*w) #expected returns
    return R
for R in returns:
    #find best volatility
    constraints = ({'type':'eq', 'fun':checkSumToOne}, 
                  {'type':'eq', 'fun': lambda w: getReturn(w) - R}) #R -> achievableReturn
    opt = minimize(minimizeMyVolatility,w0,method='SLSQP',bounds=bounds,constraints=constraints)
    #save my optimal volatility
    volatility_opt.append(opt['fun'])
```

```python
volatility_opt = [float(x) for x in volatility_opt]
volatility_opt
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,6))
plt.scatter(expectedVolatility, expectedReturn, c=sharpeRatio)
plt.xlabel('Expected Volatility')
plt.ylabel('Expected log Returns')
plt.colorbar(label='Sharpe Ratio')
plt.scatter(expectedVolatility[maxIndex],expectedReturn[maxIndex],c='red')
plt.plot(volatility_opt,returns,'--')
plt.show()
```
